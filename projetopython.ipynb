```python
# Cadastro de Funcionários com opções adicionais

print("Bem-vindo ao sistema de gerenciamento de funcionários!\n")

cadastros = {
    "funcionarios": []
}

# Funções de validação
def validar_nome(nome):
    return nome.replace(" ", "").isalpha()

def validar_idade(idade):
    return idade.isdigit() and int(idade) >= 16

def validar_salario(salario):
    try:
        return float(salario) > 0
    except ValueError:
        return False

def validar_email(email):
    return "@" in email and "." in email

# Função para cadastrar funcionário
def cadastrar_funcionario():
    print("\n--- Cadastro de Funcionário ---")

    while True:
        nome = input("Digite o nome do funcionário: ")
        if validar_nome(nome):
            break
        print("Nome inválido! Digite um nome sem números.")

    cidade = input("Digite a cidade onde o funcionário reside: ")
    estado = input("Digite o estado onde o funcionário reside: ")

    while True:
        idade = input("Digite a idade do funcionário: ")
        if validar_idade(idade):
            break
        print("Idade inválida! O funcionário deve ter 16 anos ou mais.")

    escolaridade = input("Digite a escolaridade do funcionário: ")
    cargo = input("Digite o cargo do funcionário: ")

    while True:
        salario = input("Digite o salário do funcionário: ")
        if validar_salario(salario):
            salario = float(salario)
            break
        print("Salário inválido! Digite um valor numérico maior que zero.")

    while True:
        email = input("Digite o e-mail do funcionário: ")
        if validar_email(email):
            break
        print("E-mail inválido! Digite um e-mail válido.")

    id_funcionario = len(cadastros["funcionarios"]) + 1

    funcionario = {
        "matricula": id_funcionario,
        "nome": nome,
        "cidade": cidade,
        "estado": estado,
        "idade": idade,
        "escolaridade": escolaridade,
        "cargo": cargo,
        "salario": salario,
        "email": email
    }

    cadastros["funcionarios"].append(funcionario)
    print("\nFuncionário cadastrado com sucesso!")

# Função para listar todos os funcionários
def listar_funcionarios():
    print("\n--- Lista de Funcionários ---")
    if not cadastros["funcionarios"]:
        print("Nenhum funcionário cadastrado.")
    else:
        for func in cadastros["funcionarios"]:
            print(f"ID: {func['matricula']} | Nome: {func['nome']} | Cargo: {func['cargo']} | Salário: R${func['salario']:.2f}")

# Função para buscar funcionário por nome ou código
def buscar_funcionario():
    print("\n--- Buscar Funcionário ---")
    if not cadastros["funcionarios"]:
        print("Nenhum funcionário cadastrado.")
        return

    criterio = input("Digite o nome ou o ID do funcionário: ")
    encontrado = False

    for func in cadastros["funcionarios"]:
        if criterio.isdigit() and func["matricula"] == int(criterio):
            print(f"\nFuncionário encontrado: {func}")
            encontrado = True
            break
        elif func["nome"].lower() == criterio.lower():
            print(f"\nFuncionário encontrado: {func}")
            encontrado = True
            break

    if not encontrado:
        print("Funcionário não encontrado.")

# Função para excluir funcionário
def excluir_funcionario():
    print("\n--- Excluir Funcionário ---")
    if not cadastros["funcionarios"]:
        print("Nenhum funcionário cadastrado.")
        return

    criterio = input("Digite o nome ou o ID do funcionário que deseja excluir: ")
    for i, func in enumerate(cadastros["funcionarios"]):
        if criterio.isdigit() and func["matricula"] == int(criterio):
            del cadastros["funcionarios"][i]
            print(f"\nFuncionário com ID {criterio} excluído com sucesso.")
            return
        elif func["nome"].lower() == criterio.lower():
            del cadastros["funcionarios"][i]
            print(f"\nFuncionário {criterio} excluído com sucesso.")
            return

    print("Funcionário não encontrado.")

# Menu principal
def menu():
    while True:
        print("\n--- Menu ---")
        print("1. Cadastrar Funcionário")
        print("2. Listar Funcionários")
        print("3. Buscar Funcionário")
        print("4. Excluir Funcionário")
        print("5. Sair")

        opcao = input("Escolha uma opção: ")
        if opcao == "1":
            cadastrar_funcionario()
        elif opcao == "2":
            listar_funcionarios()
        elif opcao == "3":
            buscar_funcionario()
        elif opcao == "4":
            excluir_funcionario()
        elif opcao == "5":
            print("Saindo do sistema. Até mais!")
            break
        else:
            print("Opção inválida! Tente novamente.")

# Iniciar o sistema
menu()
```
